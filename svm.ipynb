```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
```

```python
#generating dataset
np.random.seed(0)
X= np.r_[np.random.randn(20,2) - [2,2], np.random.randn(20,2) + [2,2]]
Y= [0] *20 +[1] *20
```

```python
#visualizing the dataset
plt.scatter(X[:,0],X[:,1], c=Y, cmap= plt.cm.Paired)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

```python
#training the model
clf= svm.SVC(kernel='linear')
clf.fit(X,Y)
```

```python
#plotting decision boundry
ax=plt.gca()
xlim=ax.get_xlim()
ylim=ax.get_ylim()
```

```python
#creating the grid to evaluate the model
xx=np.linspace(xlim[0], xlim[1],30)
yy=np.linspace(ylim[0], ylim[1],30)
YY,XX=np.meshgrid(yy,xx)
xy= np.vstack([XX.ravel(), YY.ravel()]).T
Z= clf.decision_function(xy).reshape(XX.shape)
```

```python
#plotting decsion boundry
ax.contour(XX,YY, Z, colors='k', levels=[-1,0,1], alpha= 0.5, linestyles=['--','-','--'])
```

```python
#plotting support vector
ax.scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1],s=100, linewidths=1, facecolors='none', edgecolors='k')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
#make prediction
np.random.seed(1)
X_test=np.random.randn(10,2)
y_pred=clf.predict(X_test)
print("predictions : ",y_pred)
```
