```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor #(for classifier and regression classs)
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

Load dataset

```python
data= load_breast_cancer()
```

```python
X= data.data
Y= data.target
```

Split the data 

```python
X_train, X_test, Y_train, Y_test =train_test_split(X,Y,test_size=0.3 , random_state=42) 
```

```python
rf= RandomForestClassifier(n_estimators=100, random_state=42) #rf cladssifier with 100 decision trees
```

train the classifier

```python
rf.fit(X_train,Y_train)
```

make predictions on test sets

```python
prediction =rf.predict(X_test)
```

```python
import sklearn
accuracy =sklearn.metrics.accuracy_score(Y_test, prediction)
print("accuracy =",accuracy)
```

```python
importance =rf.feature_importances_

sort =np.argsort(importance[: : -1])
feature_names= data.feature_names[sort]
```

```python
plt.figure(figsize=(10,6))
plt.title("feature_Importance")
plt.bar(range(X.shape[1]),importance[sort])
plt.xticks(range(X.shape[1]),feature_names, rotation=90)
plt.show()
```
