```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
height= np.array([5.1,5.3,5.4,5.6,5.9])
weight= np.array([55,62,66,68,70])
```

```python
plt.scatter(height,weight)
plt.xlabel("height")
plt.ylabel("weight")
plt.show()
```

```python
m, b = np.polyfit(height, weight, 1) # polyfit to perform linear regression
#(independent variable ,dependentr variable, degree of polynomial)
```

```python
plt.scatter(height,weight)
plt.plot(height,m*height +b , color= "green")
plt.xlabel("height")
plt.ylabel("weight")
plt.show()
```

```python
new_height= 5.5
predicted_weight=m*new_height +b
print('predicted weight:', predicted_weight)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('car_purchasing_data.csv')
data.head(5)
```

```python
X = data[['gender','age','salary','debt','worth']]
y = data['amount']
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
new_data = np.array([[1, 33,75500,15645,335845]])
predicted_price = model.predict(new_data)
print('purchasing price will:',predicted_price)
```
