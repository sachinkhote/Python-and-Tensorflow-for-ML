```python
#libraries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans 
```

```python
#generating random data points

X,_= make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
#plotting data points

plt.scatter(X[:,0], X[:,1],s=50)
plt.xlabel("001")
plt.ylabel("002")
plt.title("sample points")
plt.show()
```

```python
#cretaing K-means object with k=4
kmeans= KMeans(n_clusters=4, random_state=42)

kmeans.fit(X) #fit the data
```

```python
#getting cluster labels assigned to each data points

labels= kmeans.labels_

centers= kmeans.cluster_centers_  #get the cluster ccenter

#plot the data pouints with cluster assignment
plt.scatter(X[:,0], X[:,1], c=labels, cmap="viridis")

#plotting clustter centers
plt.scatter(centers[:,0], centers[:,1], marker= "o", c="red", s=200, alpha=0.7)
plt.xlabel("001")
plt.ylabel("002")
plt.title(" K-Means clustering")
plt.show()
```
