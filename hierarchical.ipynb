```python
#libraries 

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#generating data

np.random.seed(0)
X=np.random.rand(20,2) #20 data points in 2D space
```

```python
# performing heirarchical clustering

clustering= AgglomerativeClustering(n_clusters=None,  distance_threshold=0).fit(X)
# n cluster= None coz to let the algorithm determine the optimal number of clusters automatically.
# distance_threshold=0 indicates that the algorithm should continue merging clusters until all the data points are in a single cluster.
```

```python
link= linkage(X, method="ward")
dendrogram(link, orientation="top")
plt.xlabel("points")
plt.ylabel("distance")
plt.title("dendrogram")
plt.show()
```

```python
# Advantages 
# 1. it doesn't require specifying the number of clusters in advance, allowing for an automatic determination based on the data. 
# 2. it provides an intuitive visualization through dendrograms, enabling us to explore different clustering possibilities.
# 3. it captures hierarchical relationships which can reveal nested structures within the data.
```
